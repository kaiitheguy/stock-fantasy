```python
# Cell: Test Gemini API
import os
import requests

# Load environment variables
from dotenv import load_dotenv
load_dotenv()

# Get Gemini API key and endpoint
gemini_api_key = os.getenv("GEMINI_API_KEY")
gemini_endpoint = os.getenv("GEMINI_API_ENDPOINT", "https://api.gemini.com/v1/chat")

# Test prompt
test_prompt = "Say hello!"

# Make a request to the Gemini API
headers = {
    "Authorization": f"Bearer {gemini_api_key}",
    "Content-Type": "application/json",
}
payload = {
    "prompt": test_prompt,
    "temperature": 0.1,
}

try:
    response = requests.post(gemini_endpoint, json=payload, headers=headers)
    if response.status_code == 200:
        print("Gemini API response:", response.json())
    else:
        print(f"Gemini API error: {response.status_code} - {response.text}")
except Exception as e:
    print("Error connecting to Gemini API:", str(e))
```
